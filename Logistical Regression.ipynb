```python
import pandas as pd
```

```python
heart = pd.read_csv(r"C:\Users\Carlos\Documents\Python Projects\Logistical Regression\heart.csv")
```

```python
heart
```

We want to drop any unecessary columns

```python
heart = heart.drop(columns='Unnamed: 0')
```

```python
heart
```

Now we want to categorize our ordinal data. Let's start with the 'ChestPain' column 

```python
heart['ChestPain'] = heart['ChestPain'].astype('category').cat.codes
heart
```

Let's categorize the rest of our ordinal data

```python
heart['Thal'] = heart['Thal'].astype('category').cat.codes
heart['AHD'] = heart['AHD'].astype('category').cat.codes
heart
```

Let's check to see if there are any null values

```python
heart.isnull().sum()
```

Let's drop these null values. There are only 4 null values in the 'Ca' column. It shouldnt affect our program too much

```python
heart = heart.dropna()
```

```python
heart
```

```python
x = heart.drop(columns='AHD')
```

```python
x
```

```python
y = heart['AHD']
```

```python
y
```

Now we have to train our data. Let's separate our 'training' data from our 'testing' data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=21)
```

```python
x_train
```

```python
x_test
```

Let's scale our data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

We don't use 'fit_transform' on our testing data. We don't want our testing data to have the mean and standard deviation

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

Let's import our logistical regression library

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(random_state = 0).fit(x_train_scaled, y_train)
```

```python
log.predict(x_train_scaled)
```

Let's check the accuracy of our model

```python
log.score(x_train_scaled, y_train)
```

Both 'training' and 'testing' data should have a similar score

```python
log.score(x_test_scaled, y_test)
```
